```python
# Task 1: Load and Explore the Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Convert to pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].replace({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Display the first few rows
print("First 5 rows of the dataset:")
df.head()
```

```python
# Check data structure
print("\nDataset Info:")
df.info()

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Clean the dataset if necessary (Iris has no missing values)
# df = df.dropna()    # Not needed
```

```python
# Basic statistical summary
print("\nStatistical Summary:")
df.describe()
```

```python
# Group by species and compute mean of each numeric column
species_means = df.groupby("species").mean()

print("\nAverage Measurements per Species:")
species_means
```

```python
# Observations / Patterns
print("\nObservations:")
print("- Setosa typically has the smallest petal and sepal measurements.")
print("- Virginica generally has the largest values across most features.")
print("- Versicolor sits between Setosa and Virginica in most dimensions.")
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index, df['sepal length (cm)'])
plt.title("Trend of Sepal Length Across Dataset Index")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
species_means['petal length (cm)'].plot(kind='bar')
plt.title("Average Petal Length per Species")
plt.ylabel("Petal Length (cm)")
plt.xlabel("Species")
plt.show()

```

```python
plt.figure(figsize=(8,5))
plt.hist(df['sepal length (cm)'], bins=20, edgecolor='black')
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'])
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.show()
```

```python
print("Key Findings:")
print("1. Setosa species have significantly smaller petals compared to Versicolor and Virginica.")
print("2. Virginica species show the largest petal and sepal measurements on average.")
print("3. Sepal length appears normally distributed with no extreme outliers.")
print("4. Scatter plot indicates a strong positive correlation between petal length and sepal length.")
```
